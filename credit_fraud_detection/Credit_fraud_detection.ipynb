# Data Cleaning and Preprocessing

```python
df.drop(columns=['Loan ID','Customer ID','Months since last delinquent'],inplace=True)
```

```python
df['Years in current job'].fillna(df['Years in current job'].mode().values[0],inplace=True)
```

```python
df['Years in current job'][df['Years in current job'].str.contains('10+')]=df['Years in current job'][df['Years in current job'].str.contains('10+')].str.replace('10+','11')                                                        
```

```python
df['Years in current job'][df['Years in current job'].str.contains('< 1 year')]=df['Years in current job'][df['Years in current job'].str.contains('< 1 year')].str.replace('< 1 year','0.9')                                             
```

```python
df['Years in current job'][df['Years in current job'].str.contains('years')]=df['Years in current job'][df['Years in current job'].str.contains('years')].str.replace('years','')
```

```python
df['Years in current job'][df['Years in current job'].str.contains('year')]=df['Years in current job'][df['Years in current job'].str.contains('year')].str.replace('year','')
```

```python
df['Years in current job']=df['Years in current job'].apply(lambda x:float(x))
```

```python
c=df.select_dtypes(include='number')
c.columns
```

```python
c.drop(columns=['Number of Credit Problems','Bankruptcies','Tax Liens'],inplace=True)
for i in c.columns.values:
    q1=c[i].quantile(0.25)
    q3=c[i].quantile(0.75)
    iqr=q3-q1
    up=q3+1.5*iqr
    lo=q1-1.5*iqr
    c[i]=c[i].apply(lambda x: lo if x<lo else (up if x>up else x))
```

```python
for j in c.columns.values:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=c[j])
    plt.show()
plt.figure(figsize=(6, 4))
sns.boxplot(x=c['Credit Score'])
plt.show()
```

```python
for k in c.columns.values:
    df[k]=c[k]
    
```

# SVM

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train_sample)

x_train_sample = scaler.transform(x_train_sample)
x_test_sample = scaler.transform(x_test_sample)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],  
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'linear']
}
grid_search_svm = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    scoring='accuracy',  
    cv=3,
    n_jobs=-1 
)
grid_search_svm.fit(x_train_sample, y_train_sample)
print("Best parameters found (Grid Search):", grid_search_svm.best_params_)
```

```python
linear_svm = SVC(kernel='rbf', C=100, gamma=1)
# Train the classifier
linear_svm.fit(x_train_sample, y_train_sample)

test_predictions=linear_svm.predict(x_test_sample)
train_predictions=linear_svm.predict(x_train_sample)

print(confusion_matrix(y_test_sample, test_predictions))

print(classification_report(y_test_sample, test_predictions))

print("Train Accuracy:",accuracy_score(y_train_sample, train_predictions)*100)
print("Test Accuracy:",accuracy_score(y_test_sample, test_predictions)*100)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

param_dist = {
     'C': np.logspace(-3, 3, 5),  # Exponentially spaced values between 0.001 and 1000
    'gamma': np.logspace(-3, 3, 5),  # Same for gamma
    'kernel': ['rbf', 'poly', 'linear']  # Using RBF kernel
}

# Initialize RandomizedSearchCV
random_search_svm = RandomizedSearchCV(
    estimator=SVC(),
    param_distributions=param_dist,
    n_iter=50,  # Number of parameter settings sampled
    scoring='accuracy',  # Evaluation metric
    cv=3,  # 3-fold cross-validation
    random_state=42,
    verbose=2,
    n_jobs=-1  # Use all available cores
)

# Fit the model on training data
random_search_svm.fit(x_train_sample, y_train_sample)

# Print the best parameters
print("Best parameters found (Randomized Search):", random_search_svm.best_params_)
```
