```python
def lac(a):
    a=a.split()
    return float(a[0])*100000
def cr(a):
    a=a.split()
    return float(a[0])*10000000
```

```python
df['Amount(in rupees)'][df['Amount(in rupees)'].str.contains('Lac',case=False,na=False)]=df['Amount(in rupees)'][df['Amount(in rupees)'].str.contains('Lac',case=False,na=False)].apply(lac)
```

```python
df['Amount(in rupees)'][df['Amount(in rupees)'].str.contains('Cr',case=False,na=False)]=df['Amount(in rupees)'][df['Amount(in rupees)'].str.contains('Cr',case=False,na=False)].apply(cr)
```

```python
df['Amount(in rupees)'][df['Amount(in rupees)']=='Call for Price']=df['Amount(in rupees)'][df['Amount(in rupees)']=='Call for Price'].str.replace('Call for Price','NaN')
```

```python
df['Amount(in rupees)']=df['Amount(in rupees)'].apply(lambda x:float(x))
```

```python
df=df.dropna(subset=['Amount(in rupees)'])
```

```python
df['Price (in rupees)'].fillna(df['Price (in rupees)'].mode().values[0],inplace=True)
```

```python
df['Floor']=df['Floor'].str.split()
```

```python
df['Floor']
```

```python
l=[]
for i in df['Floor']:
    if type(i)==float:
        l.append(i)
        
    else:
        l.append(i[0])
```

```python
df['Floor']=l
```

```python
df['Floor'][df['Floor']=='Ground']=df['Floor'][df['Floor']=='Ground'].str.replace('Ground','0')
df['Floor'][df['Floor']=='Upper']=df['Floor'][df['Floor']=='Upper'].str.replace('Upper','-1')
df['Floor'][df['Floor']=='Lower']=df['Floor'][df['Floor']=='Lower'].str.replace('Lower','-2')
```

```python
df['Floor']=df['Floor'].astype('float')
```

```python
df['Floor'].fillna(df['Floor'].mode().values[0],inplace=True)
```

```python
df['Bathroom'][df['Bathroom']=='> 10']=df['Bathroom'][df['Bathroom']=='> 10'].str.replace('> 10','11')
```

```python
df['Bathroom']=df['Bathroom'].astype('float')
```

```python
df['Balcony'][df['Balcony']=='> 10']=df['Balcony'][df['Balcony']=='> 10'].str.replace('> 10','11')
```

```python
df['Balcony']=df['Balcony'].astype('float')
```

```python
df['Balcony'].fillna(df['Balcony'].mode().values[0],inplace=True)
df['Bathroom'].fillna(df['Bathroom'].mode().values[0],inplace=True)
```

```python
def convert_sqm(sqm):
    sqm=sqm.split()
    return int(sqm[0])*10.8
```

```python
def convert_sqyrd(sqm):
    sqm=sqm.split()
    return int(sqm[0])*9
```

```python
def convert_acre(sqm):
    sqm=sqm.split()
    return int(sqm[0])*43560
```

```python
def convert_ground(sqm):
    sqm=sqm.split()
    return int(sqm[0])*2400
```

```python
def convert_cent(sqm):
    sqm=sqm.split()
    return int(sqm[0])*435.6
```

```python
def convert_bigha(sqm):
    sqm=sqm.split()
    return int(sqm[0])*27000
```

```python
def convert_marla(sqm):
    sqm=sqm.split()
    return int(sqm[0])*272
```

```python
def convert_kanal(sqm):
    sqm=sqm.split()
    return int(sqm[0])*5445
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('sqm',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('sqm',case=False,na=False)].apply(convert_sqm)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('sqyrd',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('sqyrd',case=False,na=False)].apply(convert_sqyrd)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('acre',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('acre',case=False,na=False)].apply(convert_sqm)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('ground',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('ground',case=False,na=False)].apply(convert_ground)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('cent',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('cent',case=False,na=False)].apply(convert_cent)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('bigha',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('bigha',case=False,na=False)].apply(convert_bigha)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('marla',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('marla',case=False,na=False)].apply(convert_marla)
```

```python
df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('kanal',case=False,na=False)]=df['Carpet Area(sqft)'][df['Carpet Area(sqft)'].str.contains('kanal',case=False,na=False)].apply(convert_kanal)
```

```python
def label_encode(a):
    if a=='Unfurnished':
        return '0'
    elif a=='Semi-Furnished':
        return '1'
    elif a=='Furnished':
        return '2'
```

```python
df['Furnishing']=df['Furnishing'].apply(label_encode)
```

# Building Model

```python
df['Super Area'].unique()
```

```python
df.info()
```

```python
#x=df[['Price (in rupees)','location','Carpet Area','Status','Floor','Transaction','Furnishing','facing','overlooking','Bathroom','Balcony','Car Parking','Ownership','Super Area']]
x=df.iloc[:,1:]
```

```python
y=df['Amount(in rupees)']
```

```python
nd=x.select_dtypes(include='number')
```

```python
cd=x.select_dtypes(include='object')
```

```python
cd=pd.get_dummies(cd,dtype=int)
cd
```

```python
x=pd.concat([cd,nd],axis=1)
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20, random_state = 8)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# model training

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

# model testing

```python
test_predictions=model.predict(x_test)
```

```python
test_predictions
```

```python
y_test
```

```python
train_predictions=model.predict(x_train)
```

```python
y_train
```

```python
#len(train_predictions[train_predictions<0])
```

```python
print("Test Mean Squared Error: ",mean_squared_error(y_test,test_predictions))
```

```python
print("Train Mean Squared Error: ",mean_squared_error(y_train,train_predictions))
```

```python
print("Test R2 Score: ",r2_score(y_test,test_predictions))
```

```python
print("Train R2 Score: ",r2_score(y_train,train_predictions))
```

__Ridge Model__

```python
ridge_model = Ridge(alpha = 0.01)
ridge_model.fit(x_train,y_train)
```

```python
rigde_test=ridge_model.predict(x_test)
rigde_train=ridge_model.predict(x_train)
```

```python
print("Ridge Model MSE of Train: ",mean_squared_error(y_train,rigde_train))
```

```python
print("Ridge Model MSE of Test: ",mean_squared_error(y_test,rigde_test))
```

```python
print("Ridge Model Train R^2 value: ",r2_score(y_train,rigde_train)*100)
```

```python
print("Ridge Model Test R^2 value: ",r2_score(y_test,rigde_test)*100)
```

```python
Lasso_model=Lasso(alpha=0.1)
Lasso_model.fit(x_train,y_train)
```

```python
Lasso_test=Lasso_model.predict(x_test)
Lasso_train=Lasso_model.predict(x_train)
```
