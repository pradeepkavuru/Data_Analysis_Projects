# Data visualization
To uncover patterns and trends, dive deep into the dataframes using visualization techniques

```python
df.value_counts('type').plot.pie(autopct='%.2f%%')
plt.title('Type')
plt.show()
```

This pie chart shows which dominates more

```python
df[df['cast'].str.split(',')=='Akshay Kumar']
```

```python
plt.figure(figsize=(15,5))
df[df['type']=='Movie'].value_counts('director')[:15].plot.bar()
plt.title('Directors with most no.of films')
plt.ylabel('No.of Films')
plt.show()
```

```python
j=df['cast'][df['type']=='Movie'].str.split(',')
```

```python
j_exploded=j.explode('cast')
j_exploded=j_exploded.str.strip()
j_exploded.value_counts()[:60]
```

```python
j_exploded[(j_exploded=='Salman Khan') | (j_exploded=='Akshay Kumar') | (j_exploded=='Shah Rukh Khan')].value_counts().reset_index()
```

```python
plt.figure(figsize=(15,5))
j_exploded.value_counts()[:10].plot.bar()
plt.title('Actors who acted in most no.of of films')
plt.show()
```

```python
k=df['listed_in'][df['type']=='TV Show'].str.split(',')
k_exploded=k.explode()
k_exploded=k_exploded.str.strip()
k_exploded.value_counts()
```

```python
plt.figure(figsize=(15,5))
k_exploded.value_counts()[:10].plot.pie(autopct='%1.1f%%')
plt.title('Distribution of TV Shows by Genre')
plt.show()
```

```python
o=df['listed_in'][df['type']=='Movie'].str.split(',')
o_exploded=o.explode()
o_exploded=o_exploded.str.strip()
```

```python
plt.figure(figsize=(15,5))
o_exploded.value_counts()[:10].plot.bar()
plt.title('Distribution of Movie by Genre')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='year_added',data=df)
plt.xlabel('year')
plt.ylabel('Count')
plt.title('No.of Movies added per Each year')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='month_added',data=df)
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('No.of Movies added per Each month')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='day_added',data=df)
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('No.of Movies added per Each day of month')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='release_year',data=df)
plt.xticks(rotation=90)
plt.xlabel('Released Year')
plt.ylabel('Count')
plt.title('Movies released per Every year')
plt.show()
```

```python
plt.figure(figsize=(15,5))
df[df['type']=='TV Show'].value_counts('director')[:9].plot.bar()
plt.title('Directors with most TV Shows')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(15,5))
df['rating'][df['type']=='TV Show'].value_counts()[-21:-1].plot.bar()
plt.title('Distribution of TV Shows by Rating')
plt.show()
```

```python
plt.figure(figsize=(15,5))
df['rating'][df['type']=='Movie'].value_counts().plot.bar()
plt.title('Distribution of Movies by Rating')
plt.show()
```

```python
plt.figure(figsize=(15,5))
df['duration'][df['type']=='TV Show'].value_counts().plot.bar()
plt.title('Distribution of TV Shows by Duration')
plt.show()
```

```python
l=df['country'].str.split(',')
l_exploded=l.explode()
l_exploded=l_exploded.str.strip()
```

```python
plt.figure(figsize=(15,5))
l_exploded.value_counts()[0:15].plot.bar()
plt.title('Countries that produce most content')
plt.show()
```

```python
plt.figure(figsize=(15,5))
l_exploded.value_counts()[-21:-1].plot.bar()
plt.title('Countries with least movies produced')
plt.show()

```

```python
d=df['duration'][df["type"] == "Movie"].str.replace("min", '').astype("float").reset_index()
d
plt.figure(figsize=(15,5))
plt.hist(d,edgecolor='black',bins=[40,50,60,70,80,90,100,110,120])
plt.title('Distribution of Movies by Duration')
plt.xlabel('Duration')
plt.show()
```
